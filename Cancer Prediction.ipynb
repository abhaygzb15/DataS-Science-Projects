```python
# Count plot of diagnosis
plt.figure(figsize=(8, 6))
sns.countplot(x='diagnosis', data=df)
plt.title('Count of Benign and Malignant Diagnoses')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'symmetry_mean']
plt.figure(figsize=(14, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 3, i+1)
    sns.boxplot(x='diagnosis', y=feature, data=df)
    plt.title(f'{feature} by Diagnosis')
    plt.xlabel('Diagnosis')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```

## Cancer Prediction

```python
df
```

```python
df['diagnosis']=df['diagnosis'].replace({'M':1,'B':0})
df
```

```python
X=df.drop(columns=['id','diagnosis'])
Y=df['diagnosis']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train and test dataset
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,Y_train)
print("Training Score : ",model_LR.score(X_train,Y_train))
print("Test Score : ",model_LR.score(X_test,Y_test))
```

```python
y_predicted=model_LR.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_predicted)
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_predicted))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
model_RF=RandomForestClassifier()
model_RF.fit(X_train,Y_train)
print("Training Score : ",model_RF.score(X_train,Y_train))
print("Test Score : ",model_RF.score(X_test,Y_test))
```

```python
y_predicted=model_RF.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_predicted)
cm
```
